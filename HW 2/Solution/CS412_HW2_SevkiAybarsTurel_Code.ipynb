## <b> 2.a </b> [5 pts] What is the probability that a respondent is currently employed as a Programmer given they use C/C++ at work?

## <b>2.c</b> [10 pts] What is the probability that a respondent works in the Technology industry given that they earn more than 40,000 USD annually?

P(Technology) =  0.134781048097631

P(Salary > 40000) =  0.13316582914572864

P(Salary > 40000 | Technology) =  0.1939636040834443

The answer for this question is P(Technology | Salary > 40000) that equals <b> 0.19631626235399818</b>

```python
# Rename the columns in conversion_rates_df for clarity and to ensure they match the relevant columns in multiple_choice_responses_df
conversion_rates_df = conversion_rates_df.rename(columns={'originCountry': 'CompensationCurrency', 'exchangeRate': 'Rate'})

# Merge the multiple_choice_responses_df with conversion_rates_df to bring in the conversion rates
merged_df = multiple_choice_responses_df.merge(conversion_rates_df, on='CompensationCurrency', how='left')

# Create a new column 'AdjustedCompensation' by multiplying 'CompensationAmount' with 'Rate'
# Since 'CompensationAmount' might be in different formats (string with commas, etc.), we'll ensure it's converted to float
merged_df['CompensationAmount'] = pd.to_numeric(merged_df['CompensationAmount'], errors='coerce')
merged_df['AdjustedCompensation'] = merged_df['CompensationAmount'] * merged_df['Rate']

# Let's check the first few rows of the merged DataFrame to see the new 'AdjustedCompensation' column
#merged_df[['CompensationCurrency', 'CompensationAmount', 'Rate', 'AdjustedCompensation']]
```

```python
P_technology = multiple_choice_responses_df[multiple_choice_responses_df["EmployerIndustry"] == "Technology"].shape[0]/multiple_choice_responses_df.shape[0]
#print(P_technology)
```

```python
P_salary_bigger_than_40000 = merged_df[merged_df["AdjustedCompensation"] > 40000].shape[0]/multiple_choice_responses_df.shape[0]
#P_salary_bigger_than_40000
```

```python
technology_industry_mask = multiple_choice_responses_df['EmployerIndustry'] == 'Technology'

filtered_indices = multiple_choice_responses_df[technology_industry_mask].index

compensation_above_40k_mask = merged_df.loc[filtered_indices, 'AdjustedCompensation'] > 40000


num_tech_industry_above_40k = compensation_above_40k_mask.sum()

total_tech_industry = len(filtered_indices)

P_above_40k_given_tech_industry = num_tech_industry_above_40k / total_tech_industry if total_tech_industry > 0 else 0

#num_tech_industry_above_40k, total_tech_industry, P_above_40k_given_tech_industry
```

```python
print("The probability of a respondent working in the technology industry is, P(Technology) = ", P_technology)
print("The probability of a respondent earning more than $40,000 annually is, P(Salary > 40000) = ", P_salary_bigger_than_40000)
print("Given that a respondent works in the technology industry, the likelihood that they earn more than $40,000 is, P(Salary > 40000 | Technology) = ", P_above_40k_given_tech_industry)
print("The conditional probability that a respondent works in the technology industry given that they earn more than $40,000 annually is, P(Technology | Salary > 40000) = ", (P_above_40k_given_tech_industry * P_technology)/P_salary_bigger_than_40000)
```
